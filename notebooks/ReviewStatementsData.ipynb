```python
import os
import pandas as pd
from IPython.display import HTML

BASE_PATH = 'data/statements/'
```

```python
dfs = []
for quarter in os.listdir(BASE_PATH):
    dfs.append(pd.read_table(BASE_PATH + quarter + '/sub.txt'))
df = pd.concat(dfs)
df.head(2)
```
